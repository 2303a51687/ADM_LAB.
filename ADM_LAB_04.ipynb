```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/content/titanic (3).csv')
```

```python
display(df.head())
```

```python
if 'Cabin' in df.columns:
    df['Cabin'] = df['Cabin'].astype(str)
    le = LabelEncoder()
    df['Cabin_Label'] = le.fit_transform(df['Cabin'])
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
agg_data = df.groupby('Pclass').agg({'Age': 'mean', 'Fare': 'sum'})
```

```python
pivot_table = df.pivot_table(values='Age', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
print ("Aggregated Data:")
display(agg_data)
print ("Pivot Table:")
display(pivot_table)
```

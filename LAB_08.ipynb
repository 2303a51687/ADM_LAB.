```python
import numpy as np
from scipy import stats

data = [df['age'].value_counts()]


mean = np.mean(data)
print("Mean:", mean)


median = np.median(data)
print("Median:", median)


mode = stats.mode(data)
print("Mode:", mode.mode[0], "Frequency:", mode.count[0])
```
